```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Building the Geometric Brownian Motion from the derived solutio (see my notes)

```python
def geometric_Brownian_motion(S_0,mu, sigma, T, dt, n):

  paths = []

  for i in range (n):
    prices = [S_0]
    time = 0

    while(time+dt<=T):
      prices.append(prices[-1] * np.exp((mu - 0.5 * (sigma**2)) * dt + sigma * np.sqrt(dt) * np.random.normal()))

      time += dt

    if T - (time) > 0:
      prices.append(prices[-1] * np.exp((mu - 0.5 * (sigma**2)) * (T-time) + sigma * np.sqrt(T-time) * np.random.normal()))

    paths.append(prices)

  return paths
```

```python
S_0 = 100

mu = 0.08

sigma = 0.3

T = 1

dt = 1/252

n = 100
```

```python
sample_paths = geometric_Brownian_motion(S_0, mu, sigma, T, dt, n)
```

```python
for path in sample_paths:
  plt.plot(path)
plt.show()
```
