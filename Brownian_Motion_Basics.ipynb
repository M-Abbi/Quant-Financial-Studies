# Generating Brownian Motions

Let's generate 1000 different Paths and check the Brownian Motion Properties

```python
t = [0, 1, 2, 3, 4, 5]

paths=[]
n= 1000


for j in range (n):
  W = [0]
  for i in range (len(t)-1):
    W.append(np.random.normal(0,1))
  paths.append(W)
```

```python
plt.plot(t, W)
plt.show()
```

```python
for path in paths:
  plt.scatter(t,path)
```
